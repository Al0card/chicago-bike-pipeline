```python
import pandas as pd

df = pd.read_csv("../data/raw/202501-divvy-tripdata.csv", parse_dates= ['started_at', 'ended_at']) 
```

```python
df.head()
```

```python
df.columns
```

```python
colums_to_remove = ['start_lat', 'start_lng', 'end_lat', 'end_lng']
df = df.drop(columns = colums_to_remove)
```

```python
df.columns
```

```python
df["duration"] = df['ended_at'] - df['started_at']
```

```python
df.info()
```

```python
has_negative = (df["duration"] < pd.Timedelta(0)).any()
has_zero = (df["duration"] == pd.Timedelta(0)).any()
print(has_negative, has_zero)
```

```python
df["duration_seconds"] = df["duration"].dt.total_seconds().astype(int)
df.drop(columns = ["duration"])
```

```python
df.iloc[0, 2]
```

```python
df["trip_date"] = df["started_at"].dt.normalize()
df["trip_year"] = df["started_at"].dt.year
df["trip_month"] = df["started_at"].dt.month
df["trip_weekday"] = df["started_at"].dt.day_name()
df["trip_hour"] = df["started_at"].dt.hour
```
